```python
import numpy as np
import pandas as pd
import csv
import sys
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.decomposition import PCA

def analyze(plt):
   filename  = 'm3.csv'
   df = pd.read_csv(filename, usecols = [9,10,11,12,13,14] , header = 0)
   
   results = list(map(int, df['Anomaly'])) 
   print('Number of fraudulent transactions ' , sum(results))

   features = ['Airfare', 'Lodging', 'Meals', 'Other_Transportation', 'Other_Expenses']
   targets = ['Anomaly']

   x = df.loc[:, features].values
   y = df.loc[:, targets].values

   print("Number of transactions " , len(y))
   
   scaler = preprocessing.StandardScaler()
   x = scaler.fit_transform(x)

   pca = PCA(n_components=2)
   pcs = pca.fit_transform(x)
   pdf = pd.DataFrame(data = pcs,columns = ['PC 1', 'PC 2'])

   x = pdf.values.tolist()
   x = [tuple(l) for l in x]
   y = y.tolist()
   
   be = []
   fr = []
   i = 0
   while i < len(y):
        if y[i][0] == 0:
       	    be.append(x[i])
        else:
       	    fr.append(x[i])
        i += 1
   
   plt.clf()
   plt.title('PCA for ' + filename[:-4] + ' transactions')
   plt.xlabel('PC 1')
   plt.ylabel('PC 2 ')
   
   x1,x2 = zip(*be)
   plt.scatter(x1, x2,color = 'lime' , marker='o' , label = 'Normal', edgecolor = 'black', linewidth=0.2)
   
   if len(fr) > 0:
   	   x1,x2 = zip(*fr)
   	   plt.scatter(x1, x2,color = 'orangered' , marker = 'o' , label = 'Anomalous', edgecolor = 'black', linewidth=0.2)

   ax = plt.axes()
   ax.set_facecolor("white")
    
   plt.grid(True)    
   plt.grid(which='major', linestyle=':', linewidth='0.5')  
        
   fig = filename[:-4]
   plt.legend(facecolor= "white")
   plt.savefig('./' + fig +'.png', dpi = 600)

analyze(plt)
```
